```python
import pandas as pd
```

```python
df = pd.read_csv("opsys.csv")

df
```

```python
df['OpSys'] = df['OpSys'].replace(['Linux-based'], 'Linux')
df['OpSys'] = df['OpSys'].replace(['Windows Subsystem for Linux (WSL)'], 'Windows')
```

```python
df.head(20)
```

```python
df_counts = df['OpSys'].str.split(expand=True).stack().value_counts().rename_axis('OS').reset_index(name='Count')
```

```python
df_counts.head(20)
```

```python
df_counts.to_csv('osys_ec_2021.csv', index=False)
```
