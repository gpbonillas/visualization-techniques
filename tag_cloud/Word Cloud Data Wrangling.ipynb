```python
import pandas as pd
```

```python
df = pd.read_csv("tag_cloud_ds.csv")

df
```

```python
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].replace(';', ' ')

df
```

```python
df.dropna(inplace=True)

df
```

```python
df_counts = df['LanguageHaveWorkedWith'].str.split(expand=True).stack().value_counts().rename_axis('Language').reset_index(name='Count')
```

```python
df_counts.head(5)
```

```python
df_counts.to_csv('language_count_ec.csv', index=False)
```
